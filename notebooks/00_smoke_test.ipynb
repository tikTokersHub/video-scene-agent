```python
# 1. SigLIP 2 smoke test
from transformers import AutoModel, AutoProcessor
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

model = AutoModel.from_pretrained("google/siglip2-base-patch16-224").to(device)
processor = AutoProcessor.from_pretrained("google/siglip2-base-patch16-224")

print("Device:", device)
print(f"SigLIP 2 loaded: {sum(p.numel() for p in model.parameters()) / 1e6:.1f}M params")
```

```python
# 2. Chroma smoke test
import chromadb

client = chromadb.PersistentClient(path="./chroma_db")
collection = client.get_or_create_collection("smoke_test")

collection.add(
    embeddings=[[0.1, 0.2, 0.3]],
    documents=["test"],
    metadatas=[{"frame": 0}],
    ids=["test1"],
)

print(f"Chroma works: {collection.count()} item")
```

```python
# 3. LangGraph smoke test
from langgraph.graph import StateGraph, END
from typing import TypedDict

class S(TypedDict):
    msg: str

g = StateGraph(S)
g.add_node("echo", lambda s: {"msg": s["msg"] + "!"})
g.set_entry_point("echo")
g.add_edge("echo", END)

app = g.compile()
print(app.invoke({"msg": "hi"}))
```
